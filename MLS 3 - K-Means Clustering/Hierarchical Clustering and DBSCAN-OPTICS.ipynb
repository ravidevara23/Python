## Hierarchical Clustering

https://docs.scipy.org/doc/scipy/reference/cluster.hierarchy.html

```python
!pip install numpy scipy matplotlib
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Generate synthetic data
np.random.seed(42)
data = np.random.rand(10, 2)  # 10 points in 2D

# Print data
print("Data Points:\n", data)
```

```python
# Perform hierarchical clustering using the Manhattan distance
Z = linkage(data, method='average', metric='cityblock')  # 'cityblock' is another name for Manhattan distance

# Display the dendrogram
plt.figure(figsize=(10, 8))
dendrogram(Z)
plt.title("Hierarchical Clustering Dendrogram (Manhattan distance)")
plt.xlabel("Sample index")
plt.ylabel("Distance")
plt.show()
```

## DBSCAN and OPTICS


https://scikit-learn.org/1.5/modules/generated/sklearn.cluster.DBSCAN.html

DBSCAN give you the option to select the distance metrics, the default is metric='euclidean' but also you can use Manhattan (look at the code example below)


```python
!pip install scikit-learn
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN, OPTICS

# Generate some data
X, _ = make_moons(n_samples=200, noise=0.1, random_state=42)
```

```python
# Applying DBSCAN with Manhattan metric
dbscan = DBSCAN(eps=0.2, min_samples=5, metric='manhattan')
clusters = dbscan.fit_predict(X)

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', marker='o', edgecolor='k', s=50)
plt.title('DBSCAN Clustering with Manhattan Distance')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Cluster Label')
plt.show()
```

```python
# Applying OPTICS with Manhattan metric
optics = OPTICS(min_samples=5, metric='manhattan')
clusters = optics.fit_predict(X)

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', marker='o', edgecolor='k', s=50)
plt.title('OPTICS Clustering with Manhattan Distance')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Cluster Label')
plt.show()
```
