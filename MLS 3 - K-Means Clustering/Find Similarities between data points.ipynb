## Ways to find similarities between data points

Find similarities between data points is essential in tasks like clustering, recommendation systems, nearest neighbor search, and classification. Below are the common techniques to measure similarity, along with recommendations for when to use each.

### 1. Euclidean Distance


**Formula:**

$$
Distance(p,q) = \sqrt{\sum_{i=1}^n{(q_i-p_i)^2}}
$$

**Characteristics:**

  * Measures the straight-line distance between two points in n-dimensional space.
Sensitive to the scale of features.

**When to Use:**
  * Data is continuous and numeric.
  * Features have similar scales (standardization or normalization is required).

**Example Use Case:**

  * Clustering algorithms like K-Means.
  * Image processing to compare pixel values.

### 2. Manhattan Distance

**Formula:**
$$
Distance(p,q) = \sum_{i=1}^{n}{|p_i-q_i|}
$$

**Characteristics:**
  * Measures the sum of the absolute differences of their Cartesian coordinates (taxicab geometry).

**When to use:**
  * Effective in high-dimensional spaces or in urban settings where paths have to follow a grid.

**Example:**
  * In logistics and transport, optimizing travel routes within city grids.

### 3. Cosine Similarity

**Formula:**
$$
cos(\theta)=\frac{p.q}{|p|.|q|}
$$
**Characteristics:**
  * Measures the cosine of the angle between two vectors. It is not influenced by the magnitude of the vectors, focusing instead on the orientation.

**When to use:**
  * Ideal for text data represented as TF-IDF vectors, where the angle between vectors matters more than their magnitude.

**Example:**
  * Comparing text documents in natural language processing to identify how similar they are in terms of content.

### 4. Pearson Correlation Coefficient

**Characteristics:**
  * Measures the linear correlation between two datasets, normalized between -1 and 1.

**When to use:**
  * When needing to identify the linear relationship between datasets, regardless of their scale.

**Example:**
  * In finance, comparing the performance of different stocks to see how similarly they behave relative to their means.

### 5. Jaccard Similarity

**Characteristics:**
  * Measures similarity between finite sample sets and is defined as the size of the intersection divided by the size of the union of the sample sets.

**When to use:**
  * Best for comparing sets where the presence or absence of a characteristic is more important than the magnitude of measurement.

**Example:**
  * Analyzing customer buying patterns by comparing sets of products purchased by different customers.

## Example Code

```python
import numpy as np
from scipy.spatial import distance
from scipy.stats import pearsonr
from sklearn.metrics import jaccard_score

# Define points/vectors
p = np.array([1, 2, 3])
q = np.array([4, 5, 6])

# Euclidean
print("Euclidean Distance:", np.linalg.norm(p-q))

# Manhattan
manhattan_distance = distance.cityblock(p, q)
print("Manhattan Distance:", manhattan_distance)

# Cosine
cosine_similarity = np.dot(p, q) / (np.linalg.norm(p) * np.linalg.norm(q))
print("Cosine Similarity:", cosine_similarity)

# Pearson Correlation
corr, _ = pearsonr(p, q)
print("Pearson Correlation:", corr)

# Jaccard
# Note: Jaccard needs boolean or binary values, so convert first
p_bin = np.array([0, 1, 1])
q_bin = np.array([1, 1, 0])
print("Jaccard Similarity:", jaccard_score(p_bin, q_bin))
```
